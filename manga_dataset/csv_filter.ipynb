```python
# this file is for dropping dataset rows
import os
import pandas as pd 

data = pd.read_csv('metadata.csv') 

rows = len(data.index)
print(rows)

print(data.loc[[2]][['file_name']].to_string(index=False).split()[1])

#data = data.drop([0])

for x in reversed(range(rows)):
    print(x)
    name = data.loc[[x]][['file_name']].to_string(index=False).split()[1]

    isf = os.path.isfile(name)
    print(isf)
    if not (isf):
        data = data.drop(x)

print(data)
```
